# Housing Prices in King County, WA: Support Vector Regression Analysis
Goal
- Analyze Support Vector Regression predictors in relation to predictor and response variables

## Obtain Data

```python
# global imports

# matplotlib features
import matplotlib.pyplot as plt

# numpy features
import numpy as np

# scipy features
from scipy.stats import gaussian_kde
```

```python
# import stored values 
%store -r svr_test_data
```

## Visualizations of Response and Prediction
### Scatterplot of 'predictions' vs 'price'
- Predictions range from 0 to 600000 dollars, whereas actual sale prices range from 0 to 4000000 dollars
    - Predictions underestimate expensive homes
        - Expensive homes are outliers
        - The model is not trained to understand when a home may be expensive

```python
# entire scatterplot
# color by density
x = svr_test_data['price']
y = svr_test_data['predictions']
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
# scatterplot for price vs lot square footage
plt.figure(figsize = [10,10])
plt.scatter(x, y, c=z)
plt.xlabel('Dollars')
plt.ylabel('Dollars')
plt.title('Scatterplot of Predicted Sale Prices vs Actual Sale Prices')
plt.show()
```

### Scatterplot of 'abs_difference' vs 'price'
- For homes less than 500000 dollars, as price increases, the absolute prediction difference decreases
- For homes greater than 500000 dollars, as price increases, the absolute prediction difference increases

```python
# color by density
x = svr_test_data['price']
y = svr_test_data['abs_difference']
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
# scatterplot for price vs lot square footage
plt.figure(figsize = [10,10])
plt.scatter(x, y, c=z)
plt.xlabel('Dollars')
plt.ylabel('Difference in Dollars')
plt.title('Scatterplot of Absolute Prediction Difference vs Sale Price')
plt.show()
```

## Visualizations of Predictor and Predictions
### 'abs_difference' Grouped by 'location'
- Homes sold in East Urban and Seattle have the largest absolute prediction differences
    - *WashingtonHouseSales-DataVisualization* shows that these locations also have the most outliers for price
        - The model is not well trained for predicting outlier values
- Homes sold in North, South Rural, and Vashon Island have the smallest absolute prediction differences
    - *WashingtonHouseSales-DataVisualization* shows South Rural and Vashon Island have very few outliers for price

```python
# boxplot for abs_difference vs location
ax = svr_test_data.boxplot(column='abs_difference', by='location', figsize = [10,10])
ax.set_xlabel('Location')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Absolute Prediction Difference Grouped by Location')
plt.suptitle('')
plt.show()
```

### 'abs_difference' Grouped by 'waterfront'
- Homes with a waterfront typically have a larger absolute prediction difference
    - This indicates the model has not learned the effect waterfronts have on the sale price of homes

```python
# boxplot for abs_difference vs waterfront
ax = svr_test_data.boxplot(column='abs_difference', by='waterfront', figsize = [10,10])
ax.set_xlabel('Waterfront (1 is Yes, 0 is No)')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Absolute Prediction Difference Grouped by Waterfront')
plt.suptitle('')
plt.show()
```

### 'abs_difference' Grouped by 'view'

```python
# boxplot for abs_difference vs waterfront
ax = svr_test_data.boxplot(column='abs_difference', by='view', figsize = [10,10])
ax.set_xlabel('View Rating')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Sale Price Grouped by View')
plt.suptitle('')
plt.show()
```

### 'abs_difference' Grouped by 'condition'
- Aside from outliers, condition ratings of 2 and 4 have the lowest absolute prediction difference
- Aside from outliers, condition rating of 5 has the greatest absolute prediction difference

```python
# boxplot for abs_difference vs condition
ax = svr_test_data.boxplot(column='abs_difference', by='condition', figsize = [10,10])
ax.set_xlabel('Condition Rating')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Sale Price Grouped by Condition')
plt.suptitle('')
plt.show()
```

### 'abs_difference' Grouped by 'floors'
- 2.5 floors has the largest absolute prediction difference
    - In *WashingtonHouseSales-DataVisualization*, 2.5 floors also had the greatest range (excluding outliers)
    - This means the model has a harder time determining sale price for homes with 2.5 floors

```python
# boxplot for abs_difference vs floors
ax = svr_test_data.boxplot(column='abs_difference', by='floors', figsize = [10,10])
ax.set_xlabel('Number of Floors')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Sale Price Grouped by Number of Floors')
plt.suptitle('')
plt.show()
```

### 'abs_difference' Grouped by 'bedroom'
- Homes with less than 5 bedrooms have smaller absolute prediction differences than homes with 5 or more bedrooms
    - WashingtonHouseSales-DataVisualization shows there are wider ranges for prices for homes with 5 or more bedrooms
    - This means the model has a harder time determining sale price when a home has 5 or more bedrooms

```python
# boxplot for abs_difference vs bedroom
ax = svr_test_data.boxplot(column='abs_difference', by='bedroom', figsize = [10,10])
ax.set_xlabel('Number of Bedrooms')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Absolute Prediction Difference Grouped by Number of Bedrooms')
plt.suptitle('')
plt.show()
```

### 'abs_difference' Grouped by 'bathroom'
- Homes with less than 2.75 bathrooms have smaller absolute prediction differences than homes with 2.75 or more bathrooms
    - *WashingtonHouseSales-DataVisualization* shows that there are many observations of homes with less than 2.75 bathrooms, and very few observations of homes with 2.75 or more bathrooms
        - This means there is less data to train the model about how to respond to homes with 2.75 or more bathrooms
        - Thus, the absolute prediction differences are greater for those categories

```python
# boxplot for abs_difference vs bathroom
ax = svr_test_data.boxplot(column='abs_difference', by='bathroom', figsize = [10,10])
ax.set_xlabel('Number of Bathrooms')
ax.set_ylabel('Dollars')
plt.title('Boxplot of Absolute Prediction Difference Grouped by Number of Bathrooms')
plt.suptitle('')
plt.show()
```
