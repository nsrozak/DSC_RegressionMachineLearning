## Scatterplots for Price and Quantitative Predictors
### 'price' vs 'sqftLot'

```python
# color by density
x = data_pandas['sqftLot']
y = data_pandas['price']
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
# scatterplot for price vs lot square footage
plt.figure(figsize = [10,10])
plt.scatter(x, y, c=z)
plt.xlabel('Square Footage')
plt.ylabel('Dollars')
plt.title('Scatterplot of Sale Price vs Lot Size')
plt.show()
```

### 'price' vs 'sqftAbove'

```python
# color by density
x = data_pandas['sqftAbove']
y = data_pandas['price']
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
# scatterplot for price vs above ground square footage
plt.figure(figsize = [10,10])
plt.scatter(x, y, c=z)
plt.xlabel('Square Footage')
plt.ylabel('Dollars')
plt.title('Sale Price vs Above Ground Living Space')
plt.show()
```

## Time Series Plots for Price and Time Data
### 'price' vs 'yrBuilt'

```python
# group price by year built
data_built = data_pandas[['price','yrBuilt']]
grouped_built = data_built.groupby('yrBuilt')

# initiate lists
dates = []
Q1s = []
Q2s = []
Q3s = []
# get quantiles for each year built
for key, values in grouped_built:
    date = key
    prices = list(values['price'])
    Q1 = np.percentile(prices, 25)
    Q2 = np.percentile(prices, 50)
    Q3 = np.percentile(prices, 75)
    dates.append(date)
    Q1s.append(Q1)
    Q2s.append(Q2)
    Q3s.append(Q3)
    
# ticks
t = [1900,1905,1910,1915,1920,1925,1930,1935,1940,1945,1950,1955,1960,1965,1970,1975,1980,1985,
     1990,1995,2000,2005,2010,2015]
    
# time series plot of price vs year built
plt.figure(figsize=(20,10))
plt.plot(dates, Q1s)
plt.plot(dates,Q2s)
plt.plot(dates,Q3s)
plt.xlabel('Year')
plt.ylabel('Dollars')
plt.title('Time Seris Plot of Sale Price vs Year Built')
plt.xticks(t)
plt.show()
```
