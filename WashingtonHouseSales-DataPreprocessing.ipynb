## Data Cleaning
1. Create Spark dataframe from original data
2. Remove unnecessary columns
    - 'country' is redundant since all observations are from the United States
    - 'date' is problematic for predictions since observations are only from May to July
        - Want a model that can predict prices for homes sold at all times of year
        - Training data for a specific month range means the model extrapolates for dates outside of the range
        - 'date' not considered when predicting sale price to avoid this problem 
    - 'street' is problematic for predictions since observations give exact address
        - Want a model that can predict prices for any home in Washington
        - The model is overfitted if trained for specific homes
    - 'city' gives redundant location information since there is the column 'zip'
    - 'sqftLiving' gives redundant information is there are columns 'sqftAbove' and 'sqftBelow'
3. Correct 'yrRemodeled' so all observations are greater than or equal to 'yrBuilt'
    - Homes can't be remodeled until after they are built
    - Some data points are misentered so 'yrRemodeled' is less than 'yrBuilt'
    - Other data points have missing values for 'yrRemodeled'
        - This means the home was never remodeled
    - If no remodel occurs, 'yrRemodel' equals 'yrBuilt'
    - 'yrWorked' contains information about the last year construction was performed on the house
        - Information about whether a house has been remodeled or is in its original state is lost
            - I could not trust all data supplied in 'yrRemodeled' due to misentries
            - Thus, I chose to modify the column so it contains reliable information
4. Convert 'zip' data types to integer
    - 'zip' must be an integer for later column matchings
5. Add 'location' column and remove 'zip' column by joining data_spark and data_geog
    - 'zip' is a categorical variable with many levels
    - 'location' groups together levels in 'zip' to reduce the dimensionality of the dataframe
6. Create Pandas dataframe from Spark dataframe
7. Check if there are missing values and unique values in each column
8. Drop rows where data was probably misentered
    - When number of bathrooms equals 0, number of bedrooms equals 0, and price equals 0 or 7800
        - Rows with inaccurate bathrooms and bedrooms can be dropped, replaced by the mean, or replaced by the median
            - I chose to drop these rows since there are still enough observations to make a model without these rows
        - Rows with inaccurate price must be dropped since the model is being trained to predict price

```python
# schema for spark dataframe
oschema = StructType([
    StructField('date', StringType()),
    StructField('price', DoubleType()),
    StructField('bedroom', DoubleType()),
    StructField('bathroom', DoubleType()),
    StructField('sqftLiving', IntegerType()),
    StructField('sqftLot', IntegerType()),
    StructField('floors', DoubleType()),
    StructField('waterfront', IntegerType()),
    StructField('view', IntegerType()),
    StructField('condition', IntegerType()),
    StructField('sqftAbove', IntegerType()),
    StructField('sqftBelow', IntegerType()),
    StructField('yrBuilt', IntegerType()),
    StructField('yrRemodeled', IntegerType()),
    StructField('street', StringType()),
    StructField('city', StringType()),
    StructField('zip', StringType()),
    StructField('country', StringType())
])

# spark dataframe
data_spark = spark.createDataFrame(OrigData_pd, schema=oschema)
# remove unneeded columns
data_spark = data_spark.drop('street', 'country', 'city', 'date', 'sqftLiving')
# output dataframe
data_spark.show(5)
```

```python
# function that sets year remodeled to be less than or equal to year built
def yearRemodeled(made, remodeled):
    if remodeled < made:
        remodeled = made
    return remodeled

# user defined function
yearRemodeledUDF = W.udf(yearRemodeled, IntegerType())

# correct year remodeled values in data_spark
data_spark = data_spark.withColumn('yrWorked', yearRemodeledUDF('yrBuilt', 'yrRemodeled'))
# remove column with incorrect year remodeled values
data_spark = data_spark.drop('yrRemodeled')
# output dataframe
data_spark.show(5)
```

```python
# split zip column by ' '
zip_split = W.split(data_spark['zip'], ' ')
# add column 'zipcode' with zipcode to data_spark
data_spark = data_spark.withColumn('zipcode', zip_split.getItem(1))
# remove column 'zip' that contains the string zipcode and state
data_spark = data_spark.drop('zip')
# output dataframe
data_spark.show(5)
# print schema
data_spark.printSchema()
```

```python
# convert zipcode data type to integer
data_spark = data_spark.withColumn('zip', data_spark['zipcode'].cast(IntegerType()))
# remove string versions of zipcode
data_spark = data_spark.drop('zipcode')
# output dataframe
data_spark.show(5)
# print schema
data_spark.printSchema()
```

```python
# inner join data_spark and data_geog
data_spark = data_spark.join(data_geog, 'zip')
# drop duplicate columns and rows
data_spark = data_spark.drop('zip')
data_spark = data_spark.dropDuplicates()
# output dataframe
data_spark.show(5)
```

### Handling Outliers for Predictor Variables
- I chose to keep outliers for predictor variables
    - Removing outliers could overfit the model to the training dataset
    - Including outliers teaches the model to be more robust to strange values
    - This will improve model performance when faced with new data

### Handling Outliers for Response Variable
- I chose to remove the largest outliers for the response variables
    - Outliers for the response will mistrain the model if they appear in the training dataset
    - Outliers for the response will misrepresent the model's predicting abilities if they appear in the test dataset 
    
Steps for Selecting Outliers:
1. Choose a method for finding outliers
    - 'price' data is skewed right and there are many observations
        - Selection methods that require data to be normal cannot be used
            - Standard deviation and interquartile range methods
    - Clustering methods will be used to find outliers
2. Cluster the data using K Means
    - Use the elbow method to select the optimal k number of clusters
3. Graph the clusters and select outliers
    - I chose the three largest datapoints as outliers since they are far from the rest of the data

```python
# prepare data to be compatible with scipy implementation
price = data_pandas['price'].values
price = price.reshape(-1,1)
price = price.astype('float64')
```

```python
# find SSW for different values of k
sum_squared_distances = []
K = range(1,15)
for k in K:
    centroids, euc_dist_centroids = scv.kmeans(price,k)
    SSW = np.sum(euc_dist_centroids**2)
    sum_squared_distances.append(SSW)

# plot SSW for different k's and use elbow method to select k
plt.plot(K, sum_squared_distances)
plt.xlabel('k')
plt.ylabel('Total Within Cluster Sum of Squares')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# cluster the data with the optimal k=4
centroids, avg_distance = scv.kmeans(price, 4)
groups, cdist = scv.vq(price, centroids)

# plot the clusters
plt.figure(figsize=(20,6))
plt.scatter(price, np.arange(0,4548), c=groups)
plt.scatter(centroids, [2274, 2274, 2274, 2274], c='red')
plt.xlabel('Price')
plt.ylabel('Indices')
plt.title('K Means Clusters for Price')
plt.show()
```

## Machine Learning
### Prepare Data for Machine Learning
Steps to Prepare Data
1. Binary encoding of 'location'
    - 'location' is a nominal categorical variable
        - This means categories must get encoded binarily instead of numerically
2. Split data into predictors and response
    - 'price' is the response
    - All other variables are predictors
3. Normalize the predictor data so all features have equal priorities
    - The data does not need to follow a normal distribution for the algorithms I used
    - Therefore, normalizing is preferable to standardizing this data
        - Standardizing assumes the data follows a Gaussian curve
        - This assumption is not met by my data
4. Split data into train, validation, and test sets
5. Save data so it can be accessed by other programs

```python
# binary encoding of location
data = pd.get_dummies(data_pandas, columns=['location'])
data.head()
```

```python
# split data into predictors and response
X = data.drop(['price'], axis=1)
y = data[['price']]
```

```python
# convert datatypes to float
X = X.astype(float)
# normalize predictor data
values = preprocessing.MinMaxScaler().fit_transform(X)
X.loc[:,:] = values
X.head()
```

```python
# split data into train, validate, and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 123)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.1765, random_state = 123)
```
